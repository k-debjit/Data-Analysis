```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:/Users/kumar/Downloads/python/ice.csv")
df = df.set_index("Date")
df
```

```python
#shows available styling that can be done
print(plt.style.available)
```

```python
#different styles can be used
plt.style.use('grayscale')
plt.plot(df)
```

```python
#by default it will be a line plot

df.plot(kind = "line")
```

```python
#Breaking the plot
df.plot(kind = "line", subplots = True)
```

```python
#Title and labels
df.plot(kind = "line", title = "Ice cream Rating", xlabel = "Daily Ratings", ylabel = "Scores")
```

```python
#Bar plot
df.plot(kind="bar", stacked = True)
```

```python
#specific column
df["Flavor Rating"].plot(kind="bar", stacked = True)
```

```python
#Horizontal bar chart
df.plot.barh(stacked = True)
```

```python
#scatter plot. s is the size of plots and c is color
df.plot.scatter(x = "Texture Rating" , y= "Overall Rating", s = 300, c = "red")
```

```python
#Histogram. bins represents how many tick marks will be there
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = "Flavor Rating", figsize=(10,6))
```

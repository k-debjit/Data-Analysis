```python
#create or change index of a table
df = pd.read_csv(r"C:/Users/kumar/Downloads/python/population.csv", index_col = "Country")
df
```

```python
#reset index
df.reset_index(inplace=True)
#inplace creates a column of serial number
```

```python
#another way to set index but it will not save the change
df.set_index('Country')
df
```

```python
#This will save the change
df.set_index('Country', inplace = True)
df
```

```python
#Every time inplace=True add a new column of index
df.reset_index(inplace=True)
df
```

```python
#multiple index
df.set_index(['Country','Capital'], inplace = True)
df
```

```python
#sort values alphabetically
df.sort_index()
df
#more customizable sort
df.sort_index(ascending = [False,True])
```

```python
#when there are two indexes. both of the values need to be specified
df.loc['Zimbabwe','Harare']
```

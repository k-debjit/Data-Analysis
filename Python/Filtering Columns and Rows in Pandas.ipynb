```python
#Filter Rank column
df[df['Rank']<10]
```

```python
#finding specific countries in a column
find = ['Russia','United States']
df[df['Country'].isin(find)]
```

```python
#finding data that contains certain string
df[df['Country'].str.contains('es')]
```

```python
#setting index
df2 = df.set_index('Country')
df2
```

```python
#filtering by column. axis = 1 means column. axis = 0 means row
df2.filter(items = ['Capital','CCA3'], axis = 1)
df2.filter(items = ['Andorra'], axis = 0)
#using like in filter
df2.filter(like = 'desh', axis = 0)
```

```python
#sort data by values
df[df['Rank'] < 10].sort_values(by=['Rank'])
#ascending applied to both of the values
df[df['Rank'] < 10].sort_values(by=['Rank','Country'], ascending=False)
#different ascending for different values
df[df['Rank'] < 10].sort_values(by=['Rank','Country'], ascending=[False,True])
```
